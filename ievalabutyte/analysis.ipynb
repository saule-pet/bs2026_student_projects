## 6. K-Means Clustering

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

features = df[["Age","Estimated_Revenue_M_USD","Employees","Revenue_Per_Employee_K"]].copy()
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

inertias = []
K_range = range(2,8)
for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(features_scaled)
    inertias.append(km.inertia_)

plt.figure(figsize=(8,4))
plt.plot(K_range, inertias, "bo-")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("Elbow Method - Finding Optimal K", fontsize=14)
plt.tight_layout()
plt.savefig("chart_elbow.png", dpi=150)
plt.show()

km = KMeans(n_clusters=4, random_state=42, n_init=10)
df["Cluster"] = km.fit_predict(features_scaled)

cluster_names = {}
for c in range(4):
    subset = df[df["Cluster"]==c]
    avg_rev = subset["Estimated_Revenue_M_USD"].mean()
    avg_age = subset["Age"].mean()
    avg_emp = subset["Employees"].mean()
    if avg_rev > 10000: cluster_names[c] = "Legacy Giants"
    elif avg_rev > 500 and avg_age > 15: cluster_names[c] = "Established Players"
    elif avg_emp < 200: cluster_names[c] = "Emerging Startups"
    else: cluster_names[c] = "Growth Stage"

df["Cluster_Name"] = df["Cluster"].map(cluster_names)

plt.figure(figsize=(12,8))
for name, color in zip(["Legacy Giants","Established Players","Growth Stage","Emerging Startups"],["#e74c3c","#f39c12","#2ecc71","#3498db"]):
    subset = df[df["Cluster_Name"]==name]
    if len(subset) > 0:
        plt.scatter(subset["Age"], subset["Estimated_Revenue_M_USD"], s=subset["Employees"]/100, alpha=0.7, label=name, color=color, edgecolors="white")
        for _, row in subset.iterrows():
            plt.annotate(row["Company"], (row["Age"], row["Estimated_Revenue_M_USD"]), fontsize=7, alpha=0.8, ha="center", va="bottom")

plt.yscale("log")
plt.xlabel("Company Age (Years)", fontsize=12)
plt.ylabel("Revenue (USD M, log scale)", fontsize=12)
plt.title("Company Clusters: Age vs Revenue (bubble size = employees)", fontsize=14)
plt.legend(fontsize=11)
plt.tight_layout()
plt.savefig("chart_clusters.png", dpi=150)
plt.show()

print("\nCluster Summary:")
for name in sorted(df["Cluster_Name"].unique()):
    subset = df[df["Cluster_Name"]==name]
    print(f"\n{name} ({len(subset)} companies):")
    print(f"  Companies: {', '.join(subset['Company'].values)}")
    print(f"  Avg Revenue: ${subset['Estimated_Revenue_M_USD'].mean():.0f}M")
    print(f"  Avg Age: {subset['Age'].mean():.0f} years")
```
