```python
# Export-as-a-Service Market Players Analysis
# Author: Ieva Labutyte
# Date: 2026-04-14

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style
sns.set_theme(style="whitegrid")
print("Libraries loaded successfully!")
```

```python
# Create dataset of Export-as-a-Service market players
data = {
    "Company": [
        "Global-e", "eShopWorld", "Zonos", "Passport Shipping",
        "Flow Commerce", "Flexport", "Freightos", "Pitney Bowes",
        "DHL Global Forwarding", "FedEx Trade Networks",
        "Aramex", "Ship4wd", "iContainers", "Forto",
        "Shipa Freight", "BorderGuru", "Webinterpret"
    ],
    "Founded": [
        2013, 2010, 2009, 2017, 2015, 2013, 2008, 1920,
        1969, 1971, 1982, 2022, 2006, 2016, 2019, 2015, 2007
    ],
    "Headquarters": [
        "Israel", "Ireland", "USA", "USA", "USA", "USA",
        "Israel", "USA", "Germany", "USA", "UAE", "Israel",
        "Spain", "Germany", "UAE", "Germany", "Poland"
    ],
    "Service_Type": [
        "Cross-border eCommerce", "Cross-border eCommerce",
        "Duties & taxes platform", "Cross-border shipping",
        "Cross-border eCommerce", "Freight forwarding",
        "Freight marketplace", "Shipping & mailing",
        "Freight forwarding", "Freight forwarding",
        "Logistics & shipping", "Freight marketplace",
        "Freight marketplace", "Freight forwarding",
        "Freight marketplace", "Cross-border eCommerce",
        "Cross-border eCommerce"
    ],
    "Company_Size": [
        "Medium", "Medium", "Small", "Small", "Small",
        "Large", "Medium", "Large", "Large", "Large",
        "Large", "Small", "Small", "Medium", "Medium",
        "Small", "Medium"
    ],
    "Estimated_Revenue_M_USD": [
        500, 300, 50, 40, 30, 2200, 100, 3500,
        45000, 25000, 5000, 20, 15, 200, 80, 10, 60
    ]
}

df = pd.DataFrame(data)
print(f"Dataset created with {len(df)} companies")
df
```

```python
# Chart 1: Companies by Service Type
plt.figure(figsize=(10, 5))
service_counts = df["Service_Type"].value_counts()
sns.barplot(x=service_counts.values, y=service_counts.index, palette="viridis")
plt.title("Number of Companies by Service Type", fontsize=14)
plt.xlabel("Number of Companies")
plt.ylabel("")
plt.tight_layout()
plt.savefig("chart_service_types.png", dpi=150)
plt.show()
print("Chart saved as chart_service_types.png")
```

```python
# Chart 2: Companies by Headquarters Country
plt.figure(figsize=(10, 5))
hq_counts = df["Headquarters"].value_counts()
sns.barplot(x=hq_counts.values, y=hq_counts.index, palette="coolwarm")
plt.title("EaaS Companies by Country", fontsize=14)
plt.xlabel("Number of Companies")
plt.ylabel("")
plt.tight_layout()
plt.savefig("chart_countries.png", dpi=150)
plt.show()
```

```python
# Chart 3: Revenue comparison (log scale for readability)
plt.figure(figsize=(12, 6))
df_sorted = df.sort_values("Estimated_Revenue_M_USD", ascending=True)
colors = sns.color_palette("viridis", len(df_sorted))
plt.barh(df_sorted["Company"], df_sorted["Estimated_Revenue_M_USD"], color=colors)
plt.xscale("log")
plt.title("Estimated Revenue by Company (Log Scale, USD Millions)", fontsize=14)
plt.xlabel("Revenue (USD Millions, log scale)")
plt.ylabel("")
plt.tight_layout()
plt.savefig("chart_revenue.png", dpi=150)
plt.show()
```

```python
# Chart 4: Company Size Distribution
plt.figure(figsize=(7, 7))
size_counts = df["Company_Size"].value_counts()
plt.pie(size_counts.values, labels=size_counts.index, autopct="%1.0f%%",
        colors=sns.color_palette("Set2"), startangle=140, textprops={"fontsize": 13})
plt.title("Company Size Distribution", fontsize=14)
plt.tight_layout()
plt.savefig("chart_company_sizes.png", dpi=150)
plt.show()
```

```python
# Chart 5: Company Age Analysis
df["Age"] = 2026 - df["Founded"]
plt.figure(figsize=(12, 5))
df_age = df.sort_values("Age", ascending=False)
sns.barplot(x="Age", y="Company", data=df_age, palette="magma")
plt.title("Company Age (Years Since Founded)", fontsize=14)
plt.xlabel("Years")
plt.ylabel("")
plt.tight_layout()
plt.savefig("chart_company_age.png", dpi=150)
plt.show()
```

```python
# Summary Statistics
print("=" * 50)
print("EXPORT-AS-A-SERVICE MARKET SUMMARY")
print("=" * 50)
print(f"\nTotal companies analyzed: {len(df)}")
print(f"Countries represented: {df['Headquarters'].nunique()}")
print(f"Service types: {df['Service_Type'].nunique()}")
print(f"\nOldest company: {df.loc[df['Founded'].idxmin(), 'Company']} ({df['Founded'].min()})")
print(f"Newest company: {df.loc[df['Founded'].idxmax(), 'Company']} ({df['Founded'].max()})")
print(f"\nAverage company age: {df['Age'].mean():.0f} years")
print(f"Median revenue: ${df['Estimated_Revenue_M_USD'].median():.0f}M USD")
print(f"\nCompany size breakdown:")
for size, count in df["Company_Size"].value_counts().items():
    print(f"  {size}: {count} ({count/len(df)*100:.0f}%)")
```

```python
# Chart 6: Correlation heatmap of numeric variables
size_map = {"Small": 1, "Medium": 2, "Large": 3}
df["Size_Numeric"] = df["Company_Size"].map(size_map)

corr_data = df[["Founded", "Age", "Estimated_Revenue_M_USD", "Size_Numeric"]].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_data, annot=True, cmap="RdYlGn", center=0, fmt=".2f",
            xticklabels=["Founded", "Age", "Revenue", "Size"],
            yticklabels=["Founded", "Age", "Revenue", "Size"])
plt.title("Correlation Between Company Features", fontsize=14)
plt.tight_layout()
plt.savefig("chart_correlation.png", dpi=150)
plt.show()
```
